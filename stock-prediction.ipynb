```python
import yfinance as yf
import pandas as pd

# Companies to analyze
companies = ['MSFT', 'AAPL', 'GOOG']
tickers = yf.Tickers(' '.join(companies))

# Collect balance sheet data
balance_sheets_list = []

for company in companies:
    balance_sheet_df = tickers.tickers[company].balancesheet
    balance_sheet_df = balance_sheet_df.T.reset_index()
    balance_sheet_df.rename(columns={'index': 'Date'}, inplace=True)

    balance_sheet_df['Company'] = company
    balance_sheets_list.append(balance_sheet_df)

# Combine balance sheets into one DataFrame
combined_balance_sheets = pd.concat(balance_sheets_list)
combined_balance_sheets
```

```python

stock_prices_dict = {}

for company in companies:
    stock_prices = tickers.tickers[company].history(period="5y")
    stock_prices.reset_index(inplace=True)
    stock_prices['Company'] = company
    stock_prices_dict[company] = stock_prices

stock_prices_df = pd.concat(stock_prices_dict.values(), ignore_index=True)


print(stock_prices_df)
```

```python
# Ensure Date columns are timezone-naive
combined_balance_sheets['Date'] = combined_balance_sheets['Date'].dt.tz_localize(None)
stock_prices_df['Date'] = stock_prices_df['Date'].dt.tz_localize(None)

# Now merge the data
main_bal_sheet_df = pd.merge(combined_balance_sheets, stock_prices_df, on=['Company', 'Date'], how='inner').sort_values(by='Date')

print(main_bal_sheet_df)
```
